```python
import pandas as pd
import ast
```

```python
# Code LLM responses
df = pd.read_csv('output/code_llm_responses.csv')
```

```python
# Convert the dictionary-like string in the Response column into a dictionary
df['Response'] = df['Response'].apply(eval)
```

```python
# Explode the response column
df[["Query", "Response Value"]] = pd.json_normalize(df["Response"])
```

```python
df = df.drop(columns=["Response", "Query"])
df = df.rename(columns={"Response Value": "Response"})
df = df[["LLM", "Question", "Response", "Duration"]]
df.replace("\n", "", regex=True, inplace=True)
df
```

```python
df.to_csv('output/processed_code_llm_responses.csv', index=False)
```

```python
qa_with_llama1 = pd.read_csv('output/qa_llm_responses_codellama.csv')
qa_with_llama1["Response"] = qa_with_llama1["Response"].apply(eval)
```

```python
for response in qa_with_llama1["Response"]:
    for k, v in response.items():
        print(k)
```

```python
qa_with_llama1[["Result","SPARQL","Rows","Error"]] = pd.json_normalize(qa_with_llama1["Response"])
qa_with_llama1 = qa_with_llama1.drop(columns=["Response"])
qa_with_llama1 = qa_with_llama1[
    ["LLM", "Question", "SPARQL", "Rows", "Result", "Error", "Duration"]
]
qa_with_llama1.replace("\n", "", regex=True, inplace=True)
qa_with_llama1.to_csv("output/cleaned_llm_codellamaresponses.csv", index=False)
```

```python
qa_with_codestral = pd.read_csv('output/qa_llm_responses_codestral.csv')
```

```python
qa_with_codestral["Response"] = qa_with_codestral["Response"].apply(eval)
for response in qa_with_codestral["Response"]:
    for k, v in response.items():
        print(k)
```

```python
qa_with_codestral[["Result", "SPARQL", "Rows"]] = pd.json_normalize(
    qa_with_codestral["Response"]
)
qa_with_codestral = qa_with_codestral.drop(columns=["Response"])
qa_with_codestral = qa_with_codestral[
    ["LLM", "Question", "SPARQL", "Rows", "Result", "Duration"]
]
qa_with_codestral.replace("\n", "", regex=True, inplace=True)
qa_with_codestral.to_csv("output/cleaned_llm_codestralresponses.csv", index=False)
```
